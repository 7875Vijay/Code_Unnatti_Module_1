# Markers:

# Scatter Plot:

```python
# x axis:
x = [5, 6, 9, 8, 9, 7, 8, 6]
# y axis:
y = [9, 8, 7, 6, 7, 6, 3, 1]

# plotting the datapoits:
plt.scatter(x,y)

# creating the lable
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# creating the title:
plt.title("My first plot")

# showing the plotted graph:
plt.show()
```

# BarPlot:

```python
# x axis:
x = [5,6,9,8]
# y axis:
y = [9,8,7,6]

# plotting the datapoits:
plt.bar(x,y)

# creating the lable
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# creating the title:
plt.title("My first plot")

# showing the plotted graph:
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = [5, 6, 9, 8, 9, 7, 8, 6]
# y axis:
y = [9, 8, 7, 6, 7, 6, 3, 1]

plt.plot(x,y, 'g')
plt.xlabel("x")
plt.ylabel("y")

# here we set the size for ticks rotation and color values:

plt.tick_params(axis = "x", labelsize = 18, labelrotation = -60, labelcolor = 'blue')
plt.tick_params(axis = "y", labelsize = 12, labelrotation = 20, labelcolor = 'black')

plt.show()
```

```python
plt.figure(figsize = (10, 7))

# x axis:
x = [5, 6, 9, 8, 9, 7, 8, 6]
# y axis:
y = [9, 8, 7, 6, 7, 6, 3, 1]

# plotting the datapoits:
plt.plot(x)
plt.plot(y)


# creating the lable
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# creating the title:
plt.title("My first plot")

plt.legend(['blue', 'orange'], loc = 1)

# showing the plotted graph:
plt.show()
```

```python
# draw plot as fig:

fig, ax = plt.subplots()

x = np. linspace(0, 20, 1000)
ax.plot(x, np.cos(x))
ax.axis('equal')

# text anotation to the plot where it indicates maximum value of the cury

ax.annotate('local maximum', xy = (6.28, 1), xytext = (10, 4),
           arrowprops = dict(facecolor = 'black', shrink = 0.05))
ax.annotate('local minimum', xy = (5*np.pi, -1), xytext = (2, -6),
           arrowprops = dict(arrowstyle = "->", connectionstyle = 'angle3, angleA=0, angleB=-90'))
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

w = 10
h = 7
d = 100
plt.figure(figsize = (w,h), dpi = d)

#fig = plt.figure()
# Syntax for 3D projection:

ax = plt.axes(projection = '3d')

# defining all three axis:

z = np.linspace(0, 1, 100)
x = z*np.sin(25*z)
y = z*np.cos(25*z)

# plotting:

ax.plot3D(x,y,z, 'red')
ax. set_title('3d=D Line Plot')

plt.show()
```

# Pie Plot:
